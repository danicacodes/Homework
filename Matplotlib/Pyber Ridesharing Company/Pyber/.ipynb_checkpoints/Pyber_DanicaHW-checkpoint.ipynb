## Pyber Ride Sharing Analysis
***

- Urban areas have more drivers, more rides, and more fares than Suburban or Rural. However, the fare per ride in Urban areas is less than Suburban and Rural. It could be inferred that rides in Urban areas are for those going around town instead of driving further distances. 

- Moreover, there is an inverse relationship between number of rides and fare. That is, the more rides, the lower the fare. I think this is, again, due to riders taking short trips around urban areas since venues/business/apartments are closer. In Suburban or Rural areas, a lot of venues and businesses are further away, adding to the fare cost. The more mileage, the higher the fare and the longer the trip. The longer the trip, the less volume for drivers. I wish trip time was included in this data so that I could further illustrate this hypothesis. 

- Since there is significantly less drivers in both Suburban and Rural areas, there may be opportunity to attract additional drivers. In this dataset, we don't see the number of cancelled trips due to riders not being able to find a driver using the Pyber app. This is an important statistic as riders can easily toggle between multiple applications like Lyft and Uber to find the closest driver. 

- Drivers can also choose to cancel trips if the ride is too far from city center. It's worthwhile for a driver to stay within city limits even if the fare is cheaper. Said another way, a driver makes more money if he or she has six (6) ten minute rides that pay 5.00 dollars each (30.00 dollars in total) in an Urban area vs. one trip in an hour from the city to a suburban area that charges only 20. Nonetheless, once drivers dropoff a passenger in a Suburban or Rural area, they sometimes need to drive back to the city to get another ride. As a former Lyft driver, this was always an issue for me. 

- Lastly, while the data doesn't include this data point, many ride sharing apps have instances where the fares surge due to high volume and not enough drivers. This surge in fare normally happens in Urban areas where there are more ride requests, which is another incentive to stay in the city and not drive in Suburban or Rural areas. Some of my fellow Lyft drivers often drove to Washington D.C. because the "surge" was better there than in Richmond. If Pyber wants to attract more drivers in Rural or Suburban areas, perhaps they can improve their surges in these areas. 
***

```python
%matplotlib inline
# Dependencies and Setup
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

# File to Load (Remember to change these)
city_data_to_load = "data/city_data.csv"
ride_data_to_load = "data/ride_data.csv"

# Read the City and Ride Data
city_data = pd.read_csv(city_data_to_load)
ride_data = pd.read_csv(ride_data_to_load)

# Combine the data into a single dataset
pyber_df = pd.merge(ride_data, city_data, on='city', how='left')
# Display the data table for preview
pyber_df.head()
```

```python
# Provides summary of DataFrame including the data type, number of entries and columns
pyber_df.info()
```

```python
#Provides number of rows and columns
pyber_df.shape
```

## Bubble Plot of Ride Sharing Data

```python
# Obtain the x and y coordinates for each of the three city types (Urban, Suburban, Rural)
urban_type = pyber_df[pyber_df["type"] == "Urban"].groupby([pyber_df["city"]])
suburban_type = pyber_df[pyber_df["type"] == "Suburban"].groupby([pyber_df["city"]])
rural_type = pyber_df[pyber_df["type"] == "Rural"].groupby([pyber_df["city"]])

# Number of Rides (x), Average Fare (y), Driver Count per City (s), which correlates to the bubble size in the bubble chart
# Urban
x_urban = urban_type["ride_id"].count()
y_urban = urban_type["fare"].mean()
s_urban = urban_type["driver_count"].mean()

# Suburban
x_suburban = suburban_type["ride_id"].count()
y_suburban = suburban_type["fare"].mean()
s_suburban = suburban_type["driver_count"].mean()

# Rural
x_rural = rural_type["ride_id"].count()
y_rural = rural_type["fare"].mean()
s_rural = rural_type["driver_count"].mean()


# Build the scatter plots for each city types
# Added slightly darker edge color to add more depth to graph
plt.scatter(x_urban, y_urban, label = "Urban", s=s_urban * 10, color="lightcoral", edgecolor="red", alpha = 0.75, marker='o', linewidth=1.5)
plt.scatter(x_suburban, y_suburban, label = "Suburban", s=s_suburban * 10, color="skyblue", edgecolor="blue", alpha = 0.75, marker='o', linewidth=1.5)
plt.scatter(x_rural, y_rural, label = "Rural", s=s_rural * 10, color="gold", edgecolor="darkorange", alpha = 0.75, marker='o', linewidth=1.5)


# Incorporate the other graph properties
# graph grid, light grey in html hex
plt.grid(color='#BDC3C7', linewidth=.50)

# tight_layout automatically adjusts subplot params so that the subplot(s) fits in to the figure area
# https://matplotlib.org/users/tight_layout_guide.html
plt.tight_layout()

# graph labels
plt.title("Pyber Ride Sharing Data (2016)", weight='semibold')
plt.xlabel("Total Number of Rides (Per City)", weight='semibold')
plt.ylabel("Average Fare ($)", weight='semibold')


# Create a legend
legend = plt.legend(frameon= True, edgecolor='black', title="City Types", loc="upper right")

# Adjusts the bubble size in the legend so that they are all the same size
legend.legendHandles[0]._sizes = [50]
legend.legendHandles[1]._sizes = [50]
legend.legendHandles[2]._sizes = [50]

# Incorporate a text label regarding circle size
# bbox adds a rectangular box around text label
plt.text(42, 35, "Note: \nCircle size correlates with driver count per city.", fontsize = 11, weight="semibold", bbox=dict(facecolor='grey', alpha=0.04))

# Save Figure
# bbox_inches="tight" includes text when figure is saved
plt.savefig("Images\Pyber_Sharing.png", bbox_inches="tight")

# Show Plot
plt.show()
```

## Total Fares by City Type

```python
# Calculate Ride Percents
type = pyber_df.groupby('type')['type', 'fare', 'ride_id', 'driver_count']
fare_sum = type.sum()['fare']

# Build Pie Chart
labels = fare_sum.index
colors = ["gold", "skyblue", "lightcoral"]

# Explodes the third slice which is "Urban"
explode = [0, 0, 0.1]

# pctdistance changes position of labels
plt.pie(fare_sum, explode=explode, labels=labels, colors=colors, 
        autopct="%1.1f%%", startangle=140, shadow=True)

# Looks like a circle
plt.axis("equal")
plt.tight_layout()

# Chart Title
plt.suptitle('% of Total Fares by City Type', fontweight='bold')

# Save Figure
plt.savefig("Images\Total_Fares_by_City.png", bbox_inches="tight")

# Show pie chart
plt.show()
```

## Total Rides by City Type

```python
# Calculate Driver Percents
ride_sum = type.count()['ride_id']
labels = ride_sum.index

# Build Pie Charts
colors = ['gold', 'skyblue', 'lightcoral']

# Explodes the third slice which is "Urban"
explode = [0, 0, .1]

# pctdistance changes position of labels
plt.pie(ride_sum, explode=explode, labels=labels, colors=colors,
       autopct='%1.1f%%', shadow=True, startangle=150, pctdistance=0.75)

# Looks like a circle
plt.axis("equal")
plt.tight_layout()

# Chart Title
plt.suptitle('% of Total Rides by City Type', fontweight='bold')

# Save Figure
plt.savefig("Images\Total_Rides_by_City.png", bbox_inches="tight")


# Show pie chart
plt.show()
```
